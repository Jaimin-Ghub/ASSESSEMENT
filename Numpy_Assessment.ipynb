```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1) Import pandas and read in the banklist.csv file into a dataframe called banks.

data = pd.read_csv(r"C:\Users\India\Downloads\banklist.csv")
```

```python
# 2) Show the head of the dataframe.
print(data.head())
```

```python
# 3) What are the column names?

data.columns
```

```python
# 4) How many States (ST) are represented in this data set?

num_states = data["ST"].nunique()
print("Number of represented states:", num_states)
```

```python
# 5) Get a list or array of all the states in the data set.

data1 = data["ST"]
data1 = data1.tolist() 
arr_list = np.array(data1)  
print(type(arr_list)) # Check Type of Numpy Array List
print(arr_list)
```

```python
# 6) What are the top 5 states with the most failed banks?

states_count = data["ST"].value_counts()
print(states_count.head())
```

```python
# 7) What are the top 5 acquiring institutions?

states_count = data["Acquiring Institution"].value_counts()
print(states_count.head(6)[1:]) 
```

```python
# 8) How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?

# Filter by Bank name SBT "
sbt_acq = data[data["Acquiring Institution"] == "State Bank of Texas"]

total_acquir = sbt_acq.shape[0]
texas_acquir = sbt_acq[sbt_acq["ST"] == "TX"].shape[0]

# Print results
print("Total State Bank of Texas acquired banks :", total_acquir)
print("Bank acquired in Texas:", texas_acquir)
```

```python
# 9) What is the most common city in California for a bank to fail in?

california_banks = data[data["ST"] == "CA"]

# Find most common city
most_common_city = california_banks["City"].value_counts().idxmax()

print("The most common city in California for a bank to fail in :-", most_common_city)
```

```python
data2 = pd.read_csv(r"C:\Users\India\Downloads\historical_automobile_sales.csv")
print(data2.head())
```

```python
yearly_sales = data2.groupby("Year")["Automobile_Sales"].sum()

plt.figure(figsize=(10, 6))
plt.plot(yearly_sales.index, yearly_sales.values, marker = "o", linestyle = "-", color = "blue")
plt.title("Yearly Automobile Sales")
plt.xlabel("Year")
plt.ylabel("Total Automobile Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by Year, Vehicle_Type and Recession, then sum sales
sales_by_type = data2.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].sum().unstack('Vehicle_Type')
# Plot
plt.figure(figsize=(12, 7))

for column in sales_by_type.columns:
    plt.plot(sales_by_type.index, sales_by_type[column], label=column)

plt.title('Automobile Sales Trend by Vehicle Type (with Recession Periods)')
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.legend(title='Vehicle Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data2['Recession'] = data2['Recession'].astype(str)

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=data2, x="Vehicle_Type", y="Automobile_Sales", hue="Recession", estimator=sum)

plt.title("Vehicle Type vs Automobile Sales | (Recession vs Non-Recession)")
plt.xlabel("Vehicle Type")
plt.ylabel("Total Sales")
plt.legend(title="Recession", labels=["No", "Yes"]) 
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
''' 
Q 4: Now you want to compare the sales of different vehicle types during a recession and a non-recession period
Hint:  To visualize sales of different vehicles during recession and non- recession periods, you 
can use a bar chart. You will need to group Recession, Vehicle_Typefor average 
Automobile_Sales and then plot it. Make use of sns. bar plot (x=x, y=data = df))
'''

# Group karna - average Automobile Sales nikalna
grouped_data = data2.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Plot banana
plt.figure(figsize=(14, 7))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=grouped_data)

plt.title("Vehicle Type vs Average Automobile Sales | (Recession vs Non-Recession)")
plt.xlabel("Vehicle Type")
plt.ylabel("Average Automobile Sales")
plt.legend(title="Recession", labels=["No", "Yes"]) 
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
